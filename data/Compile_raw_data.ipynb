# Combine Raw Data
Combines all 17 xlsx files from the `Raw Data` folder into a single DataFrame.

```python
import pandas as pd
import os
```

```python
RAW_DATA_DIR = "Raw Data"

xlsx_files = sorted([f for f in os.listdir(RAW_DATA_DIR) if f.endswith(".xlsx")])
print(f"Found {len(xlsx_files)} files:")
for f in xlsx_files:
    print(f"  {f}")
```

```python
dfs = []

for filename in xlsx_files:
    filepath = os.path.join(RAW_DATA_DIR, filename)
    df = pd.read_excel(filepath, sheet_name=0)
    # Drop rows where Ticker is null (blank separator rows)
    df = df.dropna(subset=["Ticker"])
    # Add a source column to track which file each row came from
    df["Source"] = os.path.splitext(filename)[0]
    dfs.append(df)
    print(f"{filename}: {len(df)} rows")

combined = pd.concat(dfs, ignore_index=True)
print(f"\nTotal rows: {len(combined)}")
```

```python
futures = combined[combined["Type"] == "Future"].reset_index(drop=True)
print(f"Futures rows: {len(futures)}")
```

```python
futures_active = futures[
    (futures["Volume"] > 1000) | (futures["Open Interest"] > 1000)
].reset_index(drop=True)
print(f"Futures with Volume or Open Interest > 1000: {len(futures_active)} rows")
```
